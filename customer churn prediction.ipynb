```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("/Churn_Modelling.csv")
```

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
print(df.columns.tolist())
```

```python
df.drop(columns=['RowNumber', 'CustomerId', 'Surname'], inplace=True)
```

```python
df
```

```python
df=pd.get_dummies(df,columns=['Geography', 'Gender'],drop_first=True)
```

```python
df
```

```python
x=df.drop(columns=['Exited'])
y=df['Exited']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
seq=Sequential()
```

```python
seq.add(Dense(units=11,activation='relu'))
seq.add(Dense(units=10,activation='relu'))
seq.add(Dense(units=1,activation='sigmoid'))
```

```python
seq.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
seq.fit(x_train,y_train,batch_size=50,epochs=50,verbose=2,validation_split=0.2)
```

```python
y_pred = seq.predict(x_test_scaled)
```

```python
y_pred
```

```python
y_pred_binary = (y_pred >= 0.5).astype(int)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred_binary)
```
